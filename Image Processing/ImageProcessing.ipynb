# Reading the colored image in opencv

```python
import cv2
```

### CV2 always reads images as BGR (Blue, Green, Red)

```python
img1 = cv2.imread('CuteCat.jpeg')
img1
```

```python
img1.shape
```

### Plot above image

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
plt.show()
```

### Showing image as RGB

```python
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img_rgb
```

```python
plt.imshow(img_rgb)
plt.show()
```

### Getting RGB Channels for above image

```python
r, g, b = cv2.split(img_rgb)
```

```python
r
```

```python
r.shape
```

```python
# Red channel
plt.imshow(r, cmap='Reds')
plt.show()
```

```python
# Green Channel
plt.imshow(g, cmap='Greens')
plt.show()
```

```python
# Blue channel
plt.imshow(b, cmap='Blues')
plt.show()
```

### Reading grayscale image

```python
img_gray = cv2.imread('CuteCat.jpeg', cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
plt.show()
```
